```python
def build_h4_linear_1a_sto3g_psi4(filename):
    """
    Build a QForte 'Molecule' object for a linear H4 molecule,
    with 1 Angstrom spacing between each H atom,
    using the STO-3G basis set,
    and Psi4 as the backend.
    """
    mol_geom = [
        ('H', (0, 0, 1.0)), 
        ('H', (0, 0, 2.0)),
        ('H', (0, 0, 3.0)),
        ('H', (0, 0, 4.0))
    ]

    mol_name = "h4_linear_1a_sto3g_psi4"

    mol_options = {
        'mol_geometry': mol_geom,
        'basis': 'sto-3g', #minimal AO basis
        'charge': 0,       #neutral
        'multiplicity': 1, #singlet
        'filename': filename,
        'nroots_fci': 8  #nick
    }

    mol_build_options = {
        'num_frozen_docc': 0, #default
        'num_frozen_uocc': 0, #default

        'run_mp2': False, #nick
        'run_cisd': False, #nick
        'run_ccsd': False, #nick
        'run_fci': True, #nick

        'symmetry': 'c1', #nick
        'build_qb_ham': False, #nick
        'build_df_ham': False, #default
        'store_mo_ints': True, #nick

        'store_mo_ints_np': False, #default
        'df_icut': 1.0e-6          #default
    }

    mol = qf.system_factory(
        system_type='molecule',
        build_type='psi4',
        **mol_options,
        **mol_build_options
    )

    return mol
```

```python
mol = build_h4_linear_1a_sto3g_psi4("h4_linear_1a_sto3g_psi4")
```
